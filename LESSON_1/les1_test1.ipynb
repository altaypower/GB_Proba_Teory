Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Это сочетания

a)

```python
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

Подмножество крестей: 4/(52/4)

```python
# Подмножество крестей: 4/(52/4). Число их сочетаний:
tref = combinations((52/4), 4)
tref
```

```python
# Из вытащенных осталось 0 карт. Число их сочетаний:
rest = combinations(39, 0)
rest
```

```python
# Общее число исходов:
general = combinations(52, 4)
general
```

```python
# Вероятность:
tref * rest / general
```

б)

```python
# Подмножество 1 туза: 1/4. Число их сочетаний:
ace_1 = combinations(4, 1)
ace_1
```

```python
# Из вытащенных осталось 3 карты. Число их сочетаний:
rest_1 = combinations(48, 3)
rest_1
```

```python
# Подмножество 2 тузов: 2/4. Число их сочетаний:
ace_2 = combinations(4, 2)
ace_2
```

```python
# Из вытащенных осталось 2 карты. Число их сочетаний:
rest_2 = combinations(48, 2)
rest_2
```

```python
# Подмножество 3 тузов: 3/4. Число их сочетаний:
ace_3 = combinations(4, 3)
ace_3
```

```python
# Из вытащенных осталось 1 карта. Число их сочетаний:
rest_3 = combinations(48, 1)
rest_3
```

```python
# Подмножество 4 тузов: 4/4. Число их сочетаний:
ace_4 = combinations(4, 4)
ace_4
```

```python
# Из вытащенных осталось 0 карт. Число их сочетаний:
rest_4 = combinations(48, 0)
rest_4
```

```python
 #Число исходов, благоприятствующих нашему событию:
 favorable = ace_1 * rest_1 + ace_2 * rest_2 + ace_3 * rest_3 + ace_4 * rest_4
 favorable
```

```python
# Общее число исходов:
general = combinations(52, 4)
general
```

```python
#Вероятность:
favorable / general
```
